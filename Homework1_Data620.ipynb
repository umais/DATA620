```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()

# Create edges with different names
G.add_edge('alice', 'ben')
G.add_edge('alice', 'claire')
G.add_edge('alice', 'derek')
G.add_edge('ben', 'derek')
G.add_edge('ben', 'emma')
G.add_edge('claire', 'derek')
G.add_edge('claire', 'frank')
G.add_edge('frank', 'grace')
G.add_edge('frank', 'derek')
G.add_edge('frank', 'claire')
G.add_edge('grace', 'emma')
G.add_edge('grace', 'derek')
G.add_edge('grace', 'frank')
G.add_edge('grace', 'harry')
G.add_edge('emma', 'ben')
G.add_edge('emma', 'derek')
G.add_edge('emma', 'grace')
G.add_edge('harry', 'frank')
G.add_edge('harry', 'grace')
G.add_edge('harry', 'ian')
G.add_edge('ian', 'harry')
G.add_edge('ian', 'julia')
G.add_edge('julia', 'ian')

# Set node positions explicitly
pos = {
    'alice': (0, 1), 'ben': (0, -1), 'claire': (1, 2), 'derek': (1, 0),
    'emma': (1, -2), 'ian': (4, 0), 'grace': (2, -1), 'julia': (5, 0),
    'frank': (2, 1), 'harry': (3, 0)
}

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=800, node_color='lightcoral')

# Draw edges
nx.draw_networkx_edges(G, pos, alpha=0.6, width=4, edge_color='black')

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_family='serif', font_color='darkblue')

plt.axis('off')
plt.show()
```
